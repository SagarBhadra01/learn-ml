Arithmetic Operations

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
```

```python
#sum of two arrays
sum_arr = arr1 + arr2
sum_arr
```

```python
#subtraction of two arrays
sub_arr = arr1 - arr2
sub_arr
```

```python
#multiplication of two arrays
mul_arr = arr1 * arr2
mul_arr
```

```python
#boolean operations
bool_arr = arr1 > 3
bool_arr
```

Broadcasting

```python
arr1 + 10
```

Delete and Shallow Copy

```python
a = np.array([1,2,3,4])
a
```

```python
slice = a[1:3]
slice *10
```

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
#a@b uses matrix multiplication
a@b
```

```python
a.dot(b)
```

Stacking array

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

Splitting array

```python
np.split(a, 2)
```

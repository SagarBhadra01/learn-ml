```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
arr
```

```python
#indexing
arr[0] #first element
```

```python
arr[:2] #first two elements
```

```python
arr[:-1] #all elements except the last one
```

```python
arr[0:2] #first two elements
```

```python
arr[::-1] #reverse the array
```

Numpy indexing and slicing of matrix

```python
arr = np.arange(1,10).reshape(3,3)
arr
```

```python
arr[2] #last row
```

```python
arr[1,1] #element at row 1 and column 1
```

```python
arr[0:2,0:2] #first two rows and first two columns
```

```python
#boolean indexing
arr= np.arange(1,13)
boolean_arr = arr > 5
boolean_arr
```

```python
#filtering the array using boolean indexing
arr =arr[boolean_arr]
arr
```

1. Which team won the most matches

```python
#team won most matches
df['match_winner'].value_counts().plot(kind='bar')
plt.title('Team with Most Wins')
```

2, Toss Decision Trends

```python
sns.countplot(x='toss_decision', data=df, palette='rainbow')
```

3. Toss winner vs match winners

```python
df[df['toss_winner'] == df['match_winner']].shape[0] / df.shape[0] * 100
```

4. How do teams wins (WON BY wickets or run)

```python
sns.countplot(x='won_by', data=df)
```

Key Player parformance

1. Most player of the match

```python
count = df['player_of_the_match'].value_counts()
top_10 = count.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_10.values, y=top_10.index, palette='viridis')
plt.title('Top 10 Players of the Match')
```

2. Top scorers

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

3. 10 best bowling figure

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```

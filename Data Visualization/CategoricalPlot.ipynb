```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

Count plot

```python
#count plot is used to show the relationship between a categorical variable and a continuous variable.
sns.countplot(x="day", data=df)
```

```python
sns.countplot(x="sex", data=df, hue="smoker")
```

Bar Plot

```python
#bar plot is used to show the relationship between a categorical variable and a continuous variable.
sns.barplot(x="sex", y="total_bill", data=df)
```

```python
#bar plot with hue is used to show the relationship between a categorical variable and a continuous variable with respect to another categorical variable. palette is used to change the color of the bars. ci is used to show the confidence interval of the bars.
sns.barplot(x="sex", y="total_bill", data=df, hue="smoker", palette="Set1", ci=None)
```

Box Plot

```python
#box plot is used to show the relationship between a categorical variable and a continuous variable. It shows the distribution of the data and the outliers.
sns.boxplot(x="tip", y="day", data=df)
```

Violin PLot

```python
#violin plot is used to show the relationship between a categorical variable and a continuous variable. It shows the distribution of the data and the outliers. It also shows the kernel density estimation of the data.
sns.violinplot(x="tip", y="day", data=df)
```

Strip plot

```python
sns.stripplot(x="tip", y="day", data=df, jitter=True)
```

Swarm plot

```python
sns.swarmplot(x="tip", y="day", data=df)
```

Difference between strip plot and swarm plot is that strip plot does not show the distribution of the data and the outliers. It only shows the individual data points. Swarm plot shows the distribution of the data and the outliers. It also shows the individual data points.

```python
#violin plot with swarm plot and strip plot is used to show the relationship between a categorical variable and a continuous variable. It shows the distribution of the data and the outliers. It also shows the kernel density estimation of the data. The swarm plot and strip plot show the individual data points.
sns.violinplot(x="tip", y="day", data=df)
sns.swarmplot(x="tip", y="day", data=df)
sns.stripplot(x="tip", y="day", data=df, jitter=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the 'tips' dataset from seaborn
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# Distribution plot for 'total_bill'
#sns.displot(df['total_bill'])
# Distribution plot for 'total_bill' with KDE(meaning Kernel Density Estimate) and bins(meaning the number of bars in the histogram)
sns.displot(df['total_bill'], kde=True, bins=30)
```

```python
#two subplot total bills and tips
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['total_bill'], kde=True, bins=30, ax=axes[0])
axes[0].set_title('Distribution of Total Bill')
sns.histplot(df['tip'], kde=True, bins=30, ax=axes[1])
axes[1].set_title('Distribution of Tips')
```

```python
#join plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

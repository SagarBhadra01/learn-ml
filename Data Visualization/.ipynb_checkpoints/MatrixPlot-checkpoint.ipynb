```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Heat Map

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
#correlation matrix is used to show the relationship between two continuous variables. It shows the strength and direction of the relationship between the variables. The correlation coefficient ranges from -1 to 1. A value of 1 indicates a perfect positive correlation, a value of -1 indicates a perfect negative correlation, and a value of 0 indicates no correlation.
tipscorr = tips[["total_bill", "tip", "size"]].corr()
tipscorr
```

```python
#heatmap is used to show the relationship between two continuous variables. It shows the strength and direction of the relationship between the variables. The correlation coefficient ranges from -1 to 1. A value of 1 indicates a perfect positive correlation, a value of -1 indicates a perfect negative correlation, and a value of 0 indicates no correlation.
sns.heatmap(tipscorr.corr(), annot=True, cmap="coolwarm")
```

```python
#clustermap is used to show the relationship between two continuous variables. It shows the strength and direction of the relationship between the variables. The correlation coefficient ranges from -1 to 1. A value of 1 indicates a perfect positive correlation, a value of -1 indicates a perfect negative correlation, and a value of 0 indicates no correlation. The clustermap also shows the hierarchical clustering of the variables.
sns.clustermap(tipscorr.corr(), annot=True, cmap="coolwarm")
```

Pivot Table

```python
pvtflights = flights.pivot_table(index="month", columns="year", values="passengers")
pvtflights
```

```python
sns.heatmap(pvtflights, cmap="coolwarm")
```

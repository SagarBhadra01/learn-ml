Matplotlib Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])
# Create a line plot
plt.plot(x, y)
```

```python
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
```

```python
#grid
plt.grid()
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.grid()
```

SubPlots

```python
#subplots is used to create multiple plots in a single figure
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 3, 5, 7, 11])
y2 = np.array([1, 4, 9, 16, 25])
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x, y1) 
plt.title('Line Plot 1')
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(x, y2)
plt.title('Line Plot 2')
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

Object Oriented Way

```python
#add_axes is used to create multiple plots in a single figure with specific positions
fig = plt.figure(dpi=100)
axis1=fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [left, bottom, width, height]
plt.plot(x, y1)
plt.title('Line Plot 1')
axis2=fig.add_axes([0.2, 0.3, 0.4, 0.5]) # [left, bottom, width, height]
plt.plot(x, y2)
plt.title('Line Plot 2')
plt.show()
```

Types of plot

```python
#scatter plot
plt.scatter(x, y1, color='red', label='y1')
plt.scatter(x, y2, color='blue', label='y2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
```

```python
#histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.title('Histogram Example')
```

```python
#boxplot
data = np.random.randn(100)
plt.boxplot(data)
plt.title('Box Plot Example')
```

Saving Plots

```python
fig = plt.figure(dpi=100)
fig.add_axes([0.1, 0.1, 0.8, 0.8])
plt.plot(x, y1)
plt.title('Line Plot 1')
plt.savefig('line_plot.png') # Save the figure as a PNG file
```

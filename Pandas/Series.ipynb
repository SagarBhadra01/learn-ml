Pandas Series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_data = [10, 20, 30]
arr = np.array(my_data)
d = {'a': 10, 'b': 20, 'c': 30}
```

```python
# Creating a Series from a list
pd.Series(my_data)
```

```python
# Creating a Series from a list with custom index
pd.Series(data=my_data, index=labels)
```

```python
# Creating a Series from a numpy array
pd.Series(arr)
```

```python
#dict
pd.Series(d)
```

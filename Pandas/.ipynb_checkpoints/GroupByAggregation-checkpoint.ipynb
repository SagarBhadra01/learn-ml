```python
import numpy as np
import pandas as pd
```

GroupBy

```python
data={
    'category' : ['A', 'A', 'B', 'B', 'C', 'C'],
    'store' : ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'sales' : [100, 150, 200, 250, 300, 350],
    'Quantity' : [10, 15, 20, 25, 30, 35],
    'date' : pd.date_range('2023-01-01', periods=6, freq='D')
}
df = pd.DataFrame(data)
df
```

```python
#groupby function is used to group the data by one or more columns and perform aggregation on the grouped data
grouped = df.groupby('category')
grouped
for i, val in grouped:
    print(i)
    print(val)
```

```python
#to perform aggregation on the grouped data we can use the sum, mean, count, etc. functions
grouped_sum = grouped[['sales', 'Quantity']].sum()
grouped_sum
```

```python
#multiple line groupby
grouped_multi = df.groupby(['category', 'store'])
grouped_multi_sum = grouped_multi[['sales', 'Quantity']].sum()
grouped_multi_sum
```

Aggregation

```python
#to perform multiple aggregations on a column we can use the agg function
df['sales'].mean()
```

```python
#to perform multiple aggregations on a column we can use the agg function
df['sales'].agg(['mean', 'sum', 'max', 'min'])
```

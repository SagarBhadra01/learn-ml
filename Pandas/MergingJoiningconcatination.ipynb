```python
import numpy as np
import pandas as pd
```

Merging 2 dataframe

```python
#dataframe 1: Employee details
df1 = {
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
}
df1 = pd.DataFrame(df1)

#dataframe 2: Employee salaries
df2 = {
    'EmployeeID': [1, 2, 3, 5],
    'Salary': [50000, 60000, 55000, 70000],
    'bonus': [5000, 6000, 5500, 7000]
}
df2 = pd.DataFrame(df2)
```

```python
df1
```

```python
df2
```

```python
# Merge the two dataframes on 'EmployeeID' using an inner join
pd.merge(df1, df2, on='EmployeeID', how='inner')
```

```python
#outer join uses all the data from both the dataframes and fills in NaN for missing values
pd.merge(df1, df2, on='EmployeeID', how='outer')
```

```python
#left join uses all the data from the left dataframe and fills in NaN for missing values from the right dataframe
pd.merge(df1, df2, on='EmployeeID', how='left')
```

```python
#right join uses all the data from the right dataframe and fills in NaN for missing values from the left dataframe
pd.merge(df1, df2, on='EmployeeID', how='right')
```

Concatinatiom of 2 dataframe

```python
#concatinate the two dataframes vertically (stacking them on top of each other)

df1 = pd.DataFrame({
    'a': ['a1' , 'a2', 'a3'],
    'b': ['b1' , 'b2', 'b3'],   
    'c': ['c1' , 'c2', 'c3']
})
df2 = pd.DataFrame({
    'a': ['a4' , 'a5', 'a6'],
    'b': ['b4' , 'b5', 'b6'],   
    'c': ['c4' , 'c5', 'c6']
})
```

```python
df1
```

```python
df2
```

```python
#concat function is used to concatenate the two dataframes vertically
pd.concat([df1, df2])
```

```python
#concat function is used to concatenate the two dataframes horizontally (side by side)
pd.concat([df1, df2], axis=1)
```

Joining 2 dataframe

```python
df1 = pd.DataFrame({
    'name' : ['Alice', 'Bob', 'Charlie'],
    'age' : [25, 30, 35]
})
df2 = pd.DataFrame({
    'city' : ['New York', 'Los Angeles', 'Chicago']
})
```

```python
df1
```

```python
df2
```

```python
#join function is used to join the two dataframes on their index
df1.join(df2)
```

```python
#join function is used to join the two dataframes on their index and add suffixes to the overlapping column names
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

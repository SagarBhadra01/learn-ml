Finding Missing Data

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'a' : [1, 2, np.nan],
    'b' : [4, np.nan, 6],  
    'c' : [7, 8, 9],
    'd' : [np.nan, 11, 12]
}
df = pd.DataFrame(data)
df
```

```python
# Check for missing values
df.isna()
```

```python
# Count of missing values in each column
df.isna().sum()
```

```python
# Check if any column has missing values
df.isna().any
```

```python
# Alternative method to check for missing values
df.isnull()
```

Removing Missing data

```python
data = {
    'a' : [1, 2, np.nan],
    'b' : [4, 8, 6],  
    'c' : [7, 8, 9],
    'd' : [np.nan, 11, 12]
}
df = pd.DataFrame(data)
df
```

```python
# Drop rows with missing values
df.dropna()
```

```python
df
```

```python
# Drop rows with at least 4 non-NA values
df.dropna(thresh=4)
```

Filling the missing data

```python
df
```

```python
# Fill missing values with 0
df.fillna(0)
```

```python
df
#nothing update, for update the data fram use inplace = True
#df.fillna(0, inplace=True)
```

```python
# Fill missing values with different values for each column
values = { 'a': 10, 'b': 20, 'c': 30, 'd': 40 }
df.fillna(value=values)
```

```python
# Fill missing values with the mean of each column
df.fillna(df.mean())
```

```python
#fill missing values with the median of each column
df.fillna(df.median())
```
